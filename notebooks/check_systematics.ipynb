```python
cat = "pt350msd40"
year = "2017"
fig, ax = plt.subplots(1,1, figsize=[6,6])
for var in h.axes[1]:
    if var not in ['nominal', 'JERDown', 'JERUp']:
        continue
    h_sliced = h[{'cat' : cat, 'variation' : var, 'year' : year}]
    h_sliced.plot(label=var)
ax.set_xlim(350, 600)
ax.legend();
```
